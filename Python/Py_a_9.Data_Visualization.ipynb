# Data Visualization

```python
# Renaming pyplot object as plt

import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-talk")

x=np.arange(10)
y1=x**2
y2=2*x+3

plt.plot(x,y2,color='green')
plt.plot(x,y1,color='red')
plt.show()
```

```python
themes=plt.style.available
print(themes)
```

```python
# Graph of size 6 rows and 3 cols

plt.figure(figsize=(6,3))
plt.plot(x,y1,color='red',label="Apple",marker='*')
plt.plot(x,y2,color='green',label='Kiwi',linestyle="dashed")


plt.xlabel("time")
plt.ylabel("price")
plt.title ("Price of fruits over time")
plt.legend()
plt.show()
```

# Bar graphs

```python
plt.bar([0,1,2],[10,20,5],width=0.5)
plt.show()
```

```python
x_cord=np.array([0,1,2])*2
plt.bar(x_cord,[10,20,15],width=0.5)
plt.bar(x_cord+5,[20,10,12],width=0.5)
plt.show()
```

```python
plt.style.use("dark_background")
x_cord=np.array([0,1,2])*2

plt.bar(x_cord,[10,20,5],width=0.5,label="current year",tick_label=["gold","silver","diamond"])
plt.bar(x_cord+0.5,[20,10,4],width=0.3,label="Next year",color='Red')
plt.title("price Comparsion")

plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)

plt.legend()
plt.show()
```

# Pie Chart

```python
plt.style.use("seaborn")
subjects="Math","chem","phy","eng"
weight=[20,10,15,5]



# autopct='%1.1f%%' use for percentage calculations
plt.pie(weight,labels=subjects,explode=(0,0,0.5,0.1),autopct='%1.1f%%',shadow=True)
plt.title("subjects")
plt.show()
```

# Normal Distribution or Probabilty Distribution
## say mean of marks are 60 and varience is 5 means majority of marks is between 55 to 65
### Standard normal distribution have mean equals 0 and varience 1 i.e most of values between 1 and -1
## Xnew=(X-mean)/varience

```python
Xnew=np.random.randn(100)
sigma=5
u=60

X1=np.round(Xnew*sigma+u)
X2=np.round(Xnew*5+40)

# alpha specifies the overlap between two histograms

plt.style.use("seaborn")
plt.hist(X1,alpha=0.7,label="Phy")
plt.hist(X2,alpha=0.9,label="Maths")


plt.xlabel("Marks")
plt.title("Histogram")
plt.show()
```

## Plot length of movie title names on X-axis and freq_count on Y-axis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df=pd.read_csv("movie_metadata.csv")
df.columns
```

```python
titles=df.get('movie_title')
print(type(titles))


# It is a pandas list but we want general list hence typecasting
titles=list(df.get('movie_title'))
print(type(titles))


# Printing first 5 titles 
print(titles[:5])


# Making a dict of titles
freq_titles={}


for title in titles:
    length=len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length]=1
    else:
        freq_titles[length]+=1

        
X=np.array(list(freq_titles.keys()))
Y=np.array(list(freq_titles.values()))

plt.scatter(X,Y)
plt.xlabel("Length of movie title")
plt.ylabel("Freq count")
plt.show()
```
