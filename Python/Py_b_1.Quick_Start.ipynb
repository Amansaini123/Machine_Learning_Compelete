```python
arr=np.array([0,-9999,99,99,99,99,-123,123,1,-9999,-3,6,8,10,10,10,])
print(np.argmin(arr))
print('\n')
# Index of minimum element


# All unique elements in incresing sorted order
print(np.unique(arr))

print('\n')

# All unique elements along with their frequency
print(np.unique(arr,return_counts=True))

print('\n')

b=np.unique(arr,return_counts=True)
print(b[0])
print(b[1])

print('\n')

#Program to find the number with highest frequency

id=np.argmax(b[1])
print(b[0][id])
```
