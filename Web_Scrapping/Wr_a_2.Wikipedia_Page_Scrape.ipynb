```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup


anroid_url="https://en.wikipedia.org/wiki/Android_version_history"
anroid_data=urlopen(anroid_url)
anroid_html=anroid_data.read()
anroid_data.close()


anroid_soup=soup(anroid_html,'html.parser')
tables=anroid_soup.findAll('table',{'class':'wikitable'})
anroid_table=tables[0];


all_cols =  anroid_table.findAll('th')
col_title= [ct.text[:-1] for ct in all_cols]


table_row=[]
all_rows=anroid_table.findAll('tr')[1:]
for row in all_rows:
    curr_row=[]
    temp_row=row.findAll('td',{})
    for idx,data in enumerate(temp_row):
        curr_row.append(data.text[:-1])
    table_row.append(curr_row)
# print(table_row) 


filename='anroid_version_history.csv'
with open(filename,'w') as f:
    header_string=','.join(col_title)
    header_string+='\n'
    f.write(header_string)
    
    for row in table_row:
        row_string=""
        for w in row:
            w=w.replace(',','')
            row_string+=w+','
        row_string+='\n'
        f.write(row_string)
    
import pandas as pd
df=pd.read_csv('anroid_version_history.csv')
df.head(n=10)
```
